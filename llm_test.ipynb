```python
from curl_vertex import CurlVertex
from mylib import *
import base64, requests, json, copy, traceback
from requests.exceptions import Timeout, ReadTimeout, ConnectTimeout
import logging
from curl_vertex import CurlVertex
from mylib import LLMConfig, Message, ChatInputContentType, ChatMainLLM, ChatMainLLMName
```

```python
# Initialize Logger
logger = logging.getLogger(__name__)
logging.basicConfig(level=logging.INFO)
```

```python
def chat_with_llm(prompt: str, system_prompt: str = None):
    # Step 1: Configure LLM
    llm_config = LLMConfig({
        "model_arch": ChatMainLLM.VERTEXLLM.value,
        "model_name": ChatMainLLMName.VERTEX_GEMINI_15_PRO_2,  # Correct: Pass enum, not string
        "temperature": 0.7,
        "top_k": 10,
        "top_p": 0.95,
        "max_output_tokens": 500,
        "llm_region": "europe-west3",  # Adjust the region as necessary
        "response_mimetype": "text/plain"
    })

    # Initialize the CurlVertex instance
    curl_vertex = CurlVertex(llm_config=llm_config, logger=logger)

    # Step 2: Define chat history
    chat_history = [
        Message(
            role="user",
            message=prompt,
            message_content_type=ChatInputContentType.TEXT,
            message_uri=None,
            message_type=None
        )
    ]

    # Step 3: Generate response
    response_content = ""
    for response in curl_vertex.generate(
        instruction_prompt=system_prompt,  # Include the system prompt here
        chat_history=chat_history,
        is_streaming=True,  # Enable streaming for token-by-token responses
        return_tokens=True,  # Retrieve token usage info
        timeout=60
    ):
        response_content += response.content
        print(f"Partial Response: {response.content}")  # Print partial responses if streaming

    # Step 4: Print and return final response
    print("Final Response:")
    print(response_content)
    return response_content
```

```python
user_prompt = "Explain the significance of machine learning in healthcare."
system_prompt = "You are an expert in healthcare and technology. Provide clear and concise responses with examples when appropriate."
    
final_response = chat_with_llm(user_prompt, system_prompt)
print("LLM Output:")
print(final_response)
```
